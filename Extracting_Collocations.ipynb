```python
#reading file into a variable.
file = open (file='/users/user/positive_reviews_of_Video_Games_short.txt', mode='r')
lines = file.readlines()
document = ''
for line in lines:
    review = line.split('\t')[2]
    document += review + ('\n') 
    print(document)
file.close()    
```

contents of file is read line by line into the variable 'lines'. Initialized the variable "document" as an empty string. Looping the contents of file stored in lines. splitting by tab and 
choosing only reviews in third column to store in the variable review. then we add the contents
into the empty variable "document", creating new line for each entry using /n. Then print.

```python
#tokenizing words
from nltk import word_tokenize
words = word_tokenize(document)
print(words)
```

from NLTK, word_tokenize was imported . The reviews stored in the document is tokenized using NLTK word_tokenize and this is stored in the variable called "words" and then printed

```python
#pos_tagging tokens
from nltk import pos_tag
words_with_pos_tags = pos_tag(words)
for word, pos_tag in words_with_pos_tags:
    print(word, pos_tag)
```

from NLTK, pos_tag was imported. Next is to extract the pos_tags of each token and save it in the variable called words_with_pos_tags. next, the words and pos_tags stored in the words_with_pos_tags and printed.

```python
#extracting bigrams into list
from nltk import bigrams
bigrams_with_pos_tags = list(bigrams(words_with_pos_tags))
for first_element, second_element in bigrams_with_pos_tags:
    print(first_element, second_element)
```

first line imports bigram using nltk library. the bigram function converts the items in words_with_pos_tags into pairs of words and the list function is used to store the output in a list and store in the variable bigrams_with_pos_tags. the loop is used to iterate over the list items and print the first element and second element sequentially

```python
#iterating over the bigrams and printing the first and second items of each pair individually
for first_element, second_element in bigrams_with_pos_tags:
    first_word_bgram, first_word_ptag = first_element[0], first_element[1]
    second_word_bgram, second_word_ptag = second_element[0], second_element[1]
    print(first_word_bgram, first_word_ptag, second_word_bgram, second_word_ptag)
```

```python
#computing frequency of bigrams without filtering
def compute_frequency_of_bigrams(bigrams_with_pos_tags):
    bigrams_with_frequencies = {}
    for first_element, second_element in bigrams_with_pos_tags:
        first_word_bgram, first_word_ptag = first_element[0], first_element[1]
        second_word_bgram, second_word_ptag = second_element[0], second_element[1]
        if (first_word_bgram, second_word_bgram) in bigrams_with_frequencies:
            bigrams_with_frequencies[(first_word_bgram, second_word_bgram)] += 1
        else:
            bigrams_with_frequencies[(first_word_bgram, second_word_bgram)] = 1
    return bigrams_with_frequencies

bigrams_with_frequencies = compute_frequency_of_bigrams(bigrams_with_pos_tags)
for bigram in bigrams_with_frequencies:
    print(bigrams_with_frequencies[bigram], bigram)
```

Here, I defined a function that selects the first items of each element in bigrams_with_pos_tags and stores the first_word_bgram(bigrams) as key and second_word_bgram(frequency) is set to 1, in the empty dictionary called bigrams_with_frequencies. if pair is already in the dictionary, the integer values are increased by 1.calling function and iterating over bigrams_with_frequencies and printing bigrams and values/frequencies

```python
#sorting frequency of bigrams
import operator

topn = 40
bigrams_with_frequencies = compute_frequency_of_bigrams(bigrams_with_pos_tags)
bigrams_with_frequencies_sorted = dict(sorted(bigrams_with_frequencies.items(), key=operator.itemgetter(1), reverse=True )[:topn])
for bigram in bigrams_with_frequencies_sorted:
    print(bigrams_with_frequencies_sorted[bigram], bigram)
```

Items in bigrams_with_frequencies are sorted based on frequency in descending order and the 40 collocations with the highest frequencies are selected

A function is defined which takes in the argument words_with_pos_tags and extracts the words and stores in the dictionary called words_with_frequency. the dictionary stores words as key and integer values as frequency. if word already exists it increases by 1, if not it is set to 1.
